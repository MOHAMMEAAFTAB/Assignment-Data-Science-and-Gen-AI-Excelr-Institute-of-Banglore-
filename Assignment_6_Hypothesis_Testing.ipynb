```python
###Hypothesis Testing
import numpy as np
from scipy import stats
```

```python
# sample data
sample_mean = 3050
n = 25

# theoretical model
X = 600
mu = 1000 + 5 * X

# standard deviation
sigma = 5 * 25

print("Theoretical Mean:", mu)
print("Sigma:", sigma)
```

Null Hypothesis (H0):
The weekly operating cost follows the theoretical model.

H0 : μ = 4000

Alternative Hypothesis (H1):
The weekly operating cost is higher than the theoretical model.

H1 : μ > 4000

Calculate Test Statistic (Z)

    Formula

𝑍=(𝑥ˉ−𝜇)/(𝜎/𝑛)
Z=(xˉ−μ)/(σ/n)

```python
z = (sample_mean - mu) / (sigma / np.sqrt(n))
print("Z Statistic:", z)
```

```python
alpha = 0.05

critical_value = stats.norm.ppf(1 - alpha)

print("Critical Value:", critical_value)
```

```python
###Decision rule
if z > critical_value:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

Conclusion

The calculated Z statistic is -38.0, which is less than the critical value 1.645.

Therefore, we fail to reject the null hypothesis.

There is no strong statistical evidence to support the restaurant owners' claim that the weekly operating costs are higher than predicted by the theoretical model.

```python
import numpy as np
from scipy.stats import t, norm
```

```python
data = [1.13,1.55,1.43,0.92,1.25,1.36,1.32,0.85,1.07,1.48,1.20,1.33,1.18,1.22,1.29]

n = len(data)

mean = np.mean(data)

std = np.std(data, ddof=1)

print("Sample Mean:", mean)
print("Sample Standard Deviation:", std)
```

```python
###Sample SD
alpha = 0.01

t_value = t.ppf(1 - alpha/2, df=n-1)

margin_error = t_value * (std / np.sqrt(n))

lower = mean - margin_error
upper = mean + margin_error

print("99% CI using Sample SD:")
print(lower, upper)
```

```python
###Population SD
sigma = 0.2

z_value = norm.ppf(1 - alpha/2)

margin_error = z_value * (sigma / np.sqrt(n))

lower = mean - margin_error
upper = mean + margin_error

print("99% CI using Population SD:")
print(lower, upper)
```
