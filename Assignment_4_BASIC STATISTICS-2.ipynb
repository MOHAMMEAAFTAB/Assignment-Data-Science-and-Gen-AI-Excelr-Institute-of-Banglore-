```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
patients = pd.read_csv("Patient_Data.csv")

# show first rows
print(patients.head())

# dataset summary
patients.info()
```

```python
billing_related = patients[['PatientID','Department','Doctor','BillAmount']]

print(billing_related)
```

```python
patients_clean = patients.drop(columns=['ReceptionistID','CheckInTime'])

print(patients_clean.head())
```

```python
patients_clean = patients_clean.drop_duplicates(subset='PatientID')

print("After Removing Duplicates")
print(patients_clean)
```

```python
mean_bill = patients_clean['BillAmount'].astype(float).mean()

patients_clean['BillAmount'] = patients_clean['BillAmount'].fillna(mean_bill)

print(patients_clean)
```

```python
dept_bill = patients_clean.groupby('Department')['BillAmount'].sum()

print("Total Bill Amount per Department")
print(dept_bill)
```

```python
billing = pd.read_csv("Billing_Data.csv")

print(billing.head())
```

```python
merged_data = pd.merge(patients_clean, billing, on='PatientID')

print("Merged Dataset")
print(merged_data)
```

```python
new_patients = pd.DataFrame({
    'PatientID':[106,107],
    'Name':['Frank','Grace'],
    'Department':['Neurology','Cardiology'],
    'Doctor':['Dr. John','Dr. Smith'],
    'BillAmount':[6800,7200]
})

combined_patients = pd.concat([patients_clean, new_patients], axis=0)

print(combined_patients)
```

```python
extra_billing = pd.DataFrame({
    'InsuranceCovered':[2000,1500,2500,3000,1000],
    'FinalAmount':[3000,3500,5000,3200,4000]
})

final_dataset = pd.concat([patients_clean.reset_index(drop=True),
                           extra_billing], axis=1)

print(final_dataset)
```
