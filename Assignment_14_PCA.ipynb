```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
df = pd.read_csv("wine.csv")
df.head()
```

```python
df.info()

```

```python
df.describe()
```

```python
#missing values
df.isnull().sum()
```

```python
#Boxplot
plt.figure(figsize=(12,8))
sns.boxplot(data=df)
plt.title("Boxplot for Outliers")
plt.xticks(rotation=90)
plt.show()
```

```python
def cap_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] < lower_bound, lower_bound,
                       np.where(df[col] > upper_bound, upper_bound, df[col]))

# Apply capping to all numeric columns
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    cap_outliers(df, col)

print("Outliers capped successfully!")
```

```python
#Boxplot
plt.figure(figsize=(12,8))
sns.boxplot(data=df)
plt.title("Boxplot for Outliers")
plt.xticks(rotation=90)
plt.show()
```

```python
#heatmap distribution
df.hist(figsize=(12,10))
plt.suptitle("Feature Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
pca = PCA()
pca_data = pca.fit_transform(scaled_data)
```

```python
explained_variance = pca.explained_variance_ratio_

plt.plot(np.cumsum(explained_variance))
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Scree Plot")
plt.show()
```

```python
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
```

```python
pca_df = pd.DataFrame(data=pca_data, columns=['PC1', 'PC2'])
pca_df.head()
```

```python
kmeans_original = KMeans(n_clusters=3, random_state=42)
clusters_original = kmeans_original.fit_predict(scaled_data)
```

```python
plt.scatter(scaled_data[:,0], scaled_data[:,1], c=clusters_original)
plt.title("Clustering on Original Data")
plt.show()
```

```python
print("Silhouette Score:", silhouette_score(scaled_data, clusters_original))
print("Davies Bouldin Score:", davies_bouldin_score(scaled_data, clusters_original))
```

```python
kmeans_pca = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans_pca.fit_predict(pca_data)
```

```python
plt.scatter(pca_df['PC1'], pca_df['PC2'], c=clusters_pca)
plt.title("Clustering on PCA Data")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
print("Silhouette Score (PCA):", silhouette_score(pca_data, clusters_pca))
print("Davies Bouldin Score (PCA):", davies_bouldin_score(pca_data, clusters_pca))
```

```python
print("Original Data Clustering:")
print("Silhouette:", silhouette_score(scaled_data, clusters_original))
print("Davies Bouldin:", davies_bouldin_score(scaled_data, clusters_original))

print("\nPCA Data Clustering:")
print("Silhouette:", silhouette_score(pca_data, clusters_pca))
print("Davies Bouldin:", davies_bouldin_score(pca_data, clusters_pca))
```

## Conclusion

- PCA reduces dimensionality while retaining important information.
- Clustering on PCA data is faster and easier to visualize.
- Original data may capture more detailed structure but can be complex.

## Key Findings

- PCA improved visualization of clusters.
- Clustering performance may slightly change after PCA.
- Silhouette score helps measure cluster quality.

## Practical Implications

- Use PCA when dataset has many features.
- Use clustering directly when interpretability is important.
- PCA + Clustering is useful in real-world ML pipelines.

## Recommendation

- For large datasets → Use PCA
- For small datasets → Direct clustering is fine
