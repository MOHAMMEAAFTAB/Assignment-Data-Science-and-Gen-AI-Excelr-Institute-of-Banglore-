```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Cardiotocographic.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.columns
```

```python
###Chaecking Missing Values
df.isnull().sum()
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
df = df.apply(pd.to_numeric, errors='ignore')
```

```python
df.describe()
```

```python
df.median()
df.std()
```

The statistical summary shows mean, median, standard deviation and distribution of variables.

```python
###Detect Outlier 
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

Boxplot helps identify outliers and understand the spread of data.

```python
###Histogram_Distribution
df.hist(figsize=(12,10))
plt.show()
```

Histograms help visualize the distribution and skewness of numerical variables.

```python
###Corrrelation metrix
corr = df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

Correlation heatmap shows relationships between variables.

```python
###Scatter_plot
sns.scatterplot(x=df['LB'], y=df['AC'])
plt.title("Relationship between LB and AC")
plt.show()
```

```python
###Pair Plot (Advanced Visualization)
sns.pairplot(df)
```

```python
###violin_plot
plt.figure(figsize=(10,6))
sns.violinplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
###Pattern Recognition
df.corr()['LB'].sort_values(ascending=False)
```

Conclusion

The exploratory data analysis helped understand the distribution and relationships between fetal heart rate variables. Correlation analysis revealed relationships between accelerations, decelerations, and variability measures. Visualization techniques such as histograms, boxplots, and heatmaps helped identify patterns and potential outliers in the dataset.
